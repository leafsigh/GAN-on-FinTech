```python
import OriginalGAN as origan
import Data_Loader
import Train
import os
import pandas as pd
import numpy as np
import torch.nn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
def VisualSingleDist(real_data, fake_data):
    plt.figure(figsize=(20,16))
    plt.plot([5,6,1])
    for i in range(29):
        plt.subplot(5,6,i+1)
        sns.distplot(real_data[:,i])
        sns.distplot(fake_data[:,i])
        
def VisualBinomialDist(real_data, fake_data):
    plt.figure(figsize=(20,16))
    plt.plot([5,3,1])
    for i in range(14):
        plt.subplot(5,3,i+1)
        plt.scatter(real_data[:,int(2*i)],real_data[:,int(2*i+1)],alpha=0.3)
        plt.scatter(fake_data[:,int(2*i)],fake_data[:,int(2*i+1)],alpha=0.3)
```
